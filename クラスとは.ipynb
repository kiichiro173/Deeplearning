### クラスとは
クラスとは変数と関数（メゾッド）がセットになったもの<br>


```
class Car(object):
    #変数の定義
    def __init__(self , ini_gass , ini_speed , ini_temp):
        self.gass = ini_gass
        self.speed = ini_speed
        self.temp = ini_temp
    #関数の定義
    def run(self):
        self.gass -= 0.1
        self.speed += 0.3
    def aircon(self , target_temp):
        self.temp = target_temp 
```

```
my_car = Car(10 , 0 , 20)
```

```
my_car.speed
```

```
my_car.run()
```

```
my_car.speed
```

```
my_car.gass
```

### クラスの継承
あるクラスから新しいクラスを作成できる。<br>
クラスを継承すると変数、メゾットを受け継ぐことが出来る。<br>
super().__init__(……)これで親のクラスの初期化を行うこれを行うことによって親クラスの変数を使用する事が出来る。<br>
親のクラスにある関数を子クラスで再定義するとオーバーライドされ、子クラスで定義された方が優先される。
