○の記号で表示されている場合は、要素ごとに丸の中にある記号の処理を行っていく。
⊕　→　要素ごとに足し算を行う。<br>
⊗　→　要素ごとに掛け算を行っている。<br>

例
[2.0 , 1.0] ⊕　[x1 , x2] →　[x1 + 2.0 , x2 + 1.0]

[x1 + 2.0 , x2 + 1.0] ⊗　[3.0 , 4.0] → [3.0(x1 + 2.0 ), 4.0(x2 + 1.0)]<br>
 [3.0(x1 + 2.0 ), 4.0(x2 + 1.0)] ⇒　tanh →　 [tanh(3.0(x1 + 2.0 )), tanh(4.0(x2 + 1.0))]

□の記号で表されているのはNNの層<br>
重みづけが入るところが異なってくる。<br>
x →　σ　⇒σ(Wx + b) <br>
W:重み 　　b:バイアス　　を表している。
